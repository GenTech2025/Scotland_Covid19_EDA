#### Answer 1:
- In oder to find the difference in death rates before and after the pandemic we need data on the death trend before and after the pandemic, the dataset used below was downloaded from the dashboard at [Scotland Corona Virus Tracker](https://www.travellingtabby.com/scotland-coronavirus-tracker/)

```python
all_deaths = pd.read_csv("./CleanData/AllCausesDeathOverYears.csv")
```

```python
all_deaths.head(5)
```

```python
all_deaths.shape
```

```python
# Calculating the Summary Statistics
all_deaths.describe()
```

**Interpretation of the summary statistics table** : The average number of weekly deaths before Covid-19 pandemic is evidently less as compared to the years when the viral pandemic was in full swing or post covid years. Additionally it can be seen that the median for weekly death has been increasing gradually after the covid-19 pandemic, this could be due to the fact that a lot of people who recovered from covid had developed underlying conditions which could have affected their physical heath adversely.

```python
import plotly.express as px

# Plotting the line plot 
fig = px.line(all_deaths, x='Category', y=['2015 - 2019 Average', '2020', '2021', '2022', '2023'],
              labels={"value": "Deaths", "variable": "Year", "Category": "Week"},
              title="Weekly Deaths Over the Years From All Causes",
              hover_data={"Category": ":label Week"})  # Customizing hover label

fig.update_traces(marker=dict(size=2), line=dict(width=2), selector=dict(type='scatter', mode='lines'))
fig.update_layout(xaxis_title="Week", yaxis_title="Total Deaths")
fig.update_traces(overwrite=True, marker=dict(size=2))

# Assigning colors for each line separately
colors = ['blue', 'green', 'red', 'yellow', 'black']
for i, color in enumerate(colors):
    fig.data[i].line.color = color

fig.show()
```

### Interpretation:
- **Inspiration for make this plot was taken from  [Scotland Corona Virus Tracker](https://www.travellingtabby.com/scotland-coronavirus-tracker/)**
- The plot above shows the the average weekly death counts for the years 2015 - 2019 which gives us a rough estimate of how the average weekly deaths during the same period of time in Covid-19 pandemic years have changed.
- Its quite evident that the year 2020 which was the most devastating in terms of covid 19 pandemic showed a huge spike in the number of weekly deaths, which is mostly during the initial months of 2020, which falls in line with the unpreparedness of the the world to properly deal with the deadly virus as there were no vaccines available at that time.

## Correlation Analysis between Vaccination Rate and Death Rate

### Findings
The calculated Pearson correlation coefficient is `correlation value`. This indicates a `[weak/strong/none]` linear relationship between daily vaccination rates and daily death rates.

### Validity and Reliability Discussion
- **Data Quality**: The data used in this analysis came from `[source]`, and the reliability of our conclusions heavily depends on the accuracy and completeness of this data.
- **Methodology**: We calculated daily changes to mitigate the cumulative effect and used Pearson's correlation coefficient to measure the linear relationship.
- **Limitations**: This analysis assumes that all other conditions remain constant, which is rarely the case in real-world scenarios. Factors such as population immunity, virus variants, and public health measures can also influence death rates.
- **Potential Biases**: If data collection was biased (e.g., underreporting of deaths or vaccinations in certain regions), this would affect our findings.

### Conclusion
While the correlation coefficient provides insight into the relationship between vaccination rates and death rates, this analysis alone is not sufficient to establish causality. Further studies should incorporate more detailed data and consider other confounding variables.
